```python
import json
import pandas as pd
```

```python
rmr = json.load(open("rmr.json"))
pamr = json.load(open("Pamr.json"))
up = json.load(open("Up.json"))
Eg = json.load(open("Eg.json"))
```

```python
all_data = {}
names = ['rmr','pamr','up','Eg']
for index,data in enumerate([rmr,pamr,up,Eg]):
    name = names[index]
    for key,value in data.items():
        year = '-'.join(key.split('-')[1:])
        all_data.setdefault(name,{})[year] = value[0]
```

```python
csv = pd.DataFrame.from_dict(all_data)
```

```python
csv
```

```python
csv.plot.line(grid=True,title="Quarter CW",figsize=(20,20),fontsize=20)
```
