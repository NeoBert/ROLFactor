```python
all_algo_cws = json.load(open('all_algo_cw.json'))
bandits = ["[Rmr, 'SZ500']", "[UP, 'SZ500']", "[EG, 'SZ500']", "[Pamr, 'SZ500']", "[Olmar, 'SZ500']", "[ONS, 'SZ500']",
           "[Rmr, 'SH500']", "[UP, 'SH500']", "[EG, 'SH500']", "[Pamr, 'SH500']", "[Olmar, 'SH500']", "[ONS, 'SH500']",
           "[Rmr, 'SH50']", "[UP, 'SH50']", "[EG, 'SH50']", "[Pamr, 'SH50']", "[Olmar, 'SH50']", "[ONS, 'SH50']",
           "[Rmr, 'ZZ100']", "[UP, 'ZZ100']", "[EG, 'ZZ100']", "[Pamr, 'ZZ100']", "[Olmar, 'ZZ100']","[ONS, 'ZZ100']","bandit"]
total_cws = {}
for key,cws in all_algo_cws.items():
    total = 1
    total_cws.setdefault(bandits[int(key)],[]).append(total)
    for cw in cws:
        total = total * cw
        total_cws.setdefault(bandits[int(key)],[]).append(total)
cw = pd.DataFrame.from_dict(total_cws)
cw.plot.line(grid=True,title="Month CW",figsize=(20,20),fontsize=20)
```
